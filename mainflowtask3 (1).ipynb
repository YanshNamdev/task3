```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'C:\Drivers\diabetes.csv')
```

```python
data.head(10)
```

```python
#Line chart with SkinThickness and Glucose
plt.plot(data['SkinThickness'])
plt.plot(data['Glucose'])

#adding title to the plot
plt.title("Line plot")

#Setting the x and y labels
plt.xlabel('SkinThickness')
plt.ylabel('Glucose')

#Showing result
plt.show()
```

```python
#Bar chart and Bar plot
plt.bar(data['SkinThickness'],data['Glucose'])

#adding title to the plot
plt.title("Bar plot")

#Setting the x and y labels
plt.xlabel('SkinThickness')
plt.ylabel('Glucose')

#Showing result
plt.show()
```
